## Creating png into pixels

```python
from PIL import Image
import numpy as np

# Function to convert a PNG image to a pixel array
def png_image_to_pixel_array(image_path, size=(28, 28)):
    # Open the image file
    with Image.open(image_path) as image:
        # Convert the image to grayscale ('L' stands for luminance)
        image = image.convert('L')
        # Resize the image to the specified size
        image = image.resize(size, Image.ANTIALIAS)
        # Convert the image to a numpy array
        pixel_array = np.array(image)
        # Normalize the pixel array to the range [0, 1] if necessary
        pixel_array = pixel_array / 255.0
    return pixel_array

capitals_paths = ['A.png','B.png','C.png','D.png','E.png','F.png','G.png','H.png','I.png','J.png','K.png','L.png','M.png','N.png','O.png','P.png','Q.png','R.png','S.png','T.png','U.png','V.png','W.png','X.png','Y.png','Z.png']
len(capitals_paths)
pixels = []
flattened_pixels = []
for i,j in enumerate(capitals_paths):
    image_path = j  # Replace with the path to your PNG image
    pixels.append(png_image_to_pixel_array(image_path))

    # If you need to flatten the image for model input:
    flattened_pixels.append(pixels[i].flatten())

# `pixels` is now a 2D numpy array representing the image
# `flattened_pixels` is a 1D array of the flattened image pixels
```

```python
import matplotlib.pyplot as plt
def display_image(pixel_array):
    plt.imshow(pixel_array, cmap='gray', interpolation='nearest')
    plt.axis('off')  # Hide the axis ticks and labels
    plt.show()

# Example usage:
# Assuming you have a pixel array called 'pixels' from previous steps
display_image(pixels[2])
```
